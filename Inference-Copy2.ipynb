```python
#**Revierwer noted that I needed to check the DICOM headers, in case we need to disgard bad dicoms. Trying that
#** here

import pydicom

test_dicoms = ['test1.dcm','test2.dcm','test3.dcm','test4.dcm','test5.dcm','test6.dcm']

def check_dicom(filename): 
    # todo
    
#     print('Load file {} ...'.format(filename))
#     ds = pydicom.dcmread(filename)       
#     img = ds.pixel_array
    ds = pydicom.dcmread(filename)
    mod = ds[0x0008, 0x0060].value
    b_part_examined = ds[0x0018, 0x0015].value
    patient_position = ds[0x0018, 0x5100].value
    acceptance = 1
    
    print("Modality: " + mod)
    print("Body Part Exmained: " + b_part_examined)
    print("Patient Position: " + patient_position)
    
    if patient_position != "AP" and patient_position != "PA":
        acceptance = 0
    
    if b_part_examined != "CHEST":
        acceptance = 0
    
    if mod != "DX":
        acceptance = 0
    
    img = ds.pixel_array
    
    if acceptance:
        print("Image Accepted")
        return img
    
    if not acceptance:
        print("Image Rejected")
        return 0
    
for idx in test_dicoms:
    img = check_dicom(idx)
    print("\n")
```
